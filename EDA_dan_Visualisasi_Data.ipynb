# Import dependecies

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Import Dataset

```python
ss = pd.read_excel('/content/sample_data/SuperStore.xlsx')
ss.head()
```

# Summary Data

```python
sum_category = pd.pivot_table(ss, index=['Category'], values='Sales', aggfunc='sum')
sum_category
```

```python
sum_cat_sku = pd.pivot_table(ss, index=['Category','Sub-Category','Product_Name'], values='Sales', aggfunc=['sum','mean'])
sum_cat_sku
```

```python
ss[['Category','Product_Name','Sales']].sort_values(['Category', 'Product_Name']).head()
```

```python
ss.head()
# Granularity
# - Customer_ID
#   - Order_ID
#     - Category
#       - Sub-Category
#         - Product_ID - Product_Name
```

```python
sum_cat_sku.reset_index()
```

```python
popular_product = pd.pivot_table(
    ss,
    index='Product_Name',
    values='Order_ID',
    aggfunc='count'
).sort_values('Order_ID', ascending=False)
```

```python
popular_product.head(10).plot(kind='bar', figsize=(10,5))
```

```python
popular_product.count()
```

```python
market_share = pd.pivot_table(
    ss,
    index='Category',
    columns='Region',
    values='Sales',
    aggfunc='sum'
)
market_share
```

```python
market_share.plot(kind='bar', stacked='True', figsize=(10,5))
```

```python
market_share = market_share.unstack().reset_index()
market_share.columns = ['Region','Category','Sales']
market_share
```

```python
sns.barplot(data=market_share, x='Category', y='Sales', hue='Region')
plt.show()
```

# Line Chart

```python
ss.head(3)
```

```python
data = pd.pivot_table(
    ss,
    index=['Order_Date','Category'],
    values='Sales',
    aggfunc='sum'
).reset_index()
data.head()
```

```python
data.plot(kind='line', x='Order_Date', y='Sales', figsize=(10,5))
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(data=data, x='Order_Date', y='Sales', hue='Category')
plt.show()
```

Scatterplot

# Distplot

```python
sns.displot(data = ss[ss['Sales'] <= 1000], x='Sales', hue='Category')
plt.show
```

```python
sns.displot(data = ss[ss['Sales'] <= 1000], x='Sales', hue='Category', multiple='stack')
plt.show
```

```python
sns.displot(data = ss[ss['Sales'] <= 1000], x='Sales', hue='Category', kind='kde')
plt.show
```

```python
sns.displot(data = ss[ss['Sales'] <= 1000], x='Sales', col='Category', kind='kde')
plt.show
```

```python
sns.displot(data = ss[ss['Sales'] <= 1000], x='Sales', col='Region',row='Category', kind='kde')
plt.show
```

# Heat Map

```python
data = pd.pivot_table(
    ss,
    index='State',
    columns='Category',
    values='Sales',
    aggfunc='sum'
)
data
```

```python
sns.heatmap(data)
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(data, annot=True, cmap='Blues')
plt.show
```
